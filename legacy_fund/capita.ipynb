```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(".\csv_files\FED\WORLDGDPPERCAPITA.csv")
```

```python
cleaned = data.drop("Unnamed: 67",axis=1,errors="ignore")[["Country Name","2020","2021"]].dropna()
```

```python
cleaned["growth"] = (cleaned["2021"] - cleaned["2020"]) / cleaned["2020"]
```

```python
cleaned.sort_values("2021",ascending=True).head(30)
```

```python
plt.scatter(cleaned["2020"],cleaned["growth"])
```

```python
gdpindustry = pd.read_csv(".\csv_files\FED\WORLDGDPINDUSTRY.csv")[["Country Name","2021"]].rename(columns={"2021":"industry"})
gdpservice = pd.read_csv(".\csv_files\FED\WORLDGDPSERVICES.csv")[["Country Name","2021"]].rename(columns={"2021":"service"})
gdpagriculture = pd.read_csv(".\csv_files\FED\WORLDGDPAGRICULTURE.csv")[["Country Name","2021"]].rename(columns={"2021":"agriculture"})
gdpmanufacturing = pd.read_csv(".\csv_files\FED\WORLDGDPMANUFACTURING.csv")[["Country Name","2021"]].rename(columns={"2021":"manufacturing"})
```

```python
completed = cleaned.merge(gdpindustry,on="Country Name").merge(gdpservice,on="Country Name").merge(gdpagriculture,on="Country Name").merge(gdpmanufacturing,on="Country Name").dropna()
```

```python
completed.sort_values("2021",ascending=True).tail(30)
```
