```python
from transformer.financial_transformer import FinancialTransformer
from database.market import Market
from database.sec import SEC
import pandas as pd
from tqdm import tqdm 
from datetime import datetime
```

```python
sec = SEC()
market = Market()
```

```python
market.connect()
ciks = market.retrieve("sp500")["CIK"]
market.disconnect()
```

```python
sec.connect()
subs = sec.retrieve("subs")
adshs = sec.retrieve_adshs()
sec.disconnect()
```

```python
existing_adshs = adshs["adsh"].unique()
# existing_adshs = []
```

```python
existing_adshs
```

```python
for cik in ciks:
    print(cik,len(subs[subs["cik"]==int(cik)]))
```

```python
sec.connect()
for i in tqdm(range(ciks.index.size)):
    try:
        cik = int(ciks[i])
        values = subs[(subs["cik"]==cik) & (~subs["adsh"].isin(existing_adshs))]
        all_adshs = values["adsh"].unique()
        for adsh in all_adshs:
            try:
                if adsh not in existing_adshs:
                    num = sec.retrieve_num_data(adsh)
                    table = FinancialTransformer.transform(values,num)
                    sec.store("filings",table)
                else:
                    continue
            except Exception as e:
                print(adsh,num)
                print(str(e))
    except Exception as e:
        print(str(e))
sec.disconnect()
```
