```python
from strats.speculation import Speculation
from modeler_strats.modeler_strat_fact import ModelerStratFactory as modelstratfact
```

```python
financial = Speculation(4)
```

```python
start_year = 2023
end_year = 2024
```

```python
model_strats = [
                "universal",
                "industry",
               ]
```

```python
financial.pull_included_columns()
financial.pull_factors()
```

```python
data = financial.pull_training_data()
```

```python
data
```

```python
for model_strat in model_strats:
    modeler_class = modelstratfact.build_modeler_strat(model_strat)
    financial.db.connect()
    for year in range(start_year,end_year):
        training_set = financial.training_set(data,year)
        prediction_set = financial.prediction_set(data,year)
        stuff = modeler_class.model(training_set,prediction_set,financial.factors)
        stuff = financial.prediction_clean(stuff)
        financial.db.store(f"{modeler_class.name}_sim",stuff)
    financial.db.disconnect()
```
