```python
import pandas as pd
from database.market import Market
from database.adatabase import ADatabase
from processor.processor import Processor as p
from datetime import datetime
from tqdm import tqdm
import numpy as np
from dotenv import load_dotenv
import os
load_dotenv()
import openai
openai.api_key = os.getenv("OPENAI")
```

```python
market = Market()
spec = ADatabase("opennai")
```

```python
market.connect()
sp500 = market.retrieve("sp500")
market.disconnect()
sp500 = sp500.rename(columns={"Symbol":"ticker"})
```

```python
factors = [str(x) for x in range(14)]
```

```python
included_columns=[
    "year",
    "week",
    "ticker",
    "adjclose",
    "y"
]
```

```python
included_columns.extend([str(x) for x in range(14)])
```

```python
training_sets = []
```

```python
for ticker in sp500["ticker"].unique():
    try:
        prices = market.retrieve_ticker_prices("prices",ticker)
        prices = p.column_date_processing(prices)
        prices["year"] = [x.year for x in prices["date"]]
        prices["quarter"] = [x.quarter for x in prices["date"]]
        ticker_data = prices[prices["ticker"]==ticker]
        ticker_data.sort_values("date",ascending=True,inplace=True)
        ticker_data["adjclose"] = [float(x) for x in ticker_data["adjclose"]]
        ticker_data = ticker_data.groupby(["year","week"]).mean().reset_index()
        for i in range(14):
            ticker_data[str(i)] = ticker_data["adjclose"].shift(i)
        ticker_data["d1"] = ticker_data["adjclose"].pct_change(periods=1)
        ticker_data["d2"] = ticker_data["d1"].pct_change(periods=1)
        ticker_data["d3"] = ticker_data["d2"].pct_change(periods=1)
        ticker_data.dropna(inplace=True)
        ticker_data["ticker"] = ticker
        ticker_data["y"] = ticker_data["adjclose"].shift(-1)
        ticker_data = ticker_data.replace([np.inf, -np.inf], np.nan).dropna()
        ticker_data.dropna(inplace=True)
        ticker_data = ticker_data[included_columns]
        training_sets.append(ticker_data)
    except:
        continue
```

```python
data = pd.concat(training_sets)
```

```python
data = data.dropna()
```

```python
prediction_data = data[data["year"]>=2020]
```

```python
prediction_data["prompt"] = [str(" ").join([str(row[1][str(x)]) for x in range(14)]) for row in prediction_data.iterrows()]
```

```python
prediction_data["prompt"].iloc[0]
```

```python
spec.connect()
predictions = []
for row in prediction_data.iloc[:1].iterrows():
    prompt = row[1]["prompt"]
    print(prompt)
    response = openai.Completion.create(model="text-davinci-003", prompt=prompt, temperature=0, max_tokens=7)
    print(response)
    predictions.append(response)
spec.disconnect()
prediction_data["prediction"] = predictions
```
