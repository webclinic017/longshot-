```python
import pandas as pd
from database.market import Market
from processor.processor import Processor as p
from tqdm import tqdm
```

```python
market = Market()
```

```python
market.connect()
prices = market.retrieve("prices")
market.disconnect()
```

```python
prices = p.column_date_processing(prices)
prices["year"] = [x.year for x in prices["date"]]
```

```python
pt = pd.pivot_table(prices,index="date",values="adjclose",columns="ticker").reset_index()
```

```python
pt["year"] = [x.year for x in pt["date"]]
```

```python
hee_hee = pt.corr().reset_index()
```

```python
market.connect()
for training_year in tqdm(range(1,8)):
    for year in range(2021,2023):
        training = pt[(pt["year"]<year) & (pt["year"]>=year-training_year)]
        final = training.corr().reset_index()
        final["year"] = year
        final["training_year"] = training_year
        market.store("price_corr",final)
market.disconnect()
```
